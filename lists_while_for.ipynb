# –¶–∏–∫–ª for

—Ü–∏–∫–ª for

–û–¥–Ω–æ –∏–∑ –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤ –∫–æ–º–ø—å—é—Ç–µ—Ä–æ–≤ –ø–µ—Ä–µ–¥ –ª—é–¥—å–º–∏ - —Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å –ø–æ–≤—Ç–æ—Ä—è—Ç—å –æ–¥–Ω–∏ –∏ —Ç–µ –∂–µ –¥–µ–π—Å—Ç–≤–∏—è –º–Ω–æ–≥–æ–∫—Ä–∞—Ç–Ω–æ, –±—ã—Å—Ç—Ä–æ –∏ —Å–æ–≤—Å–µ–º –Ω–µ —É—Ç–æ–º–ª—è—è—Å—å üíª

–í Python —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –¥–≤–µ –æ—Å–Ω–æ–≤–Ω—ã—Ö —Ä–∞–∑–Ω–æ–≤–∏–¥–Ω–æ—Å—Ç–∏ —Ü–∏–∫–ª–æ–≤:

—Ü–∏–∫–ª—ã, –ø–æ–≤—Ç–æ—Ä—è—é—â–∏–µ—Å—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–∑ (for, —Å—á–µ—Ç–Ω—ã–µ —Ü–∏–∫–ª—ã, counting loops);
—Ü–∏–∫–ª—ã, –ø–æ–≤—Ç–æ—Ä—è—é—â–∏–µ—Å—è –¥–æ –Ω–∞—Å—Ç—É–ø–ª–µ–Ω–∏—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–≥–æ —Å–æ–±—ã—Ç–∏—è (while, —É—Å–ª–æ–≤–Ω—ã–µ —Ü–∏–∫–ª—ã, conditional loops).
–¶–∏–∫–ª for –∑–∞–º–µ—á–∞—Ç–µ–ª—å–Ω–æ —Ä–∞–±–æ—Ç–∞–µ—Ç, –µ—Å–ª–∏ –º—ã –∑–∞—Ä–∞–Ω–µ–µ –∑–Ω–∞–µ–º, —Å–∫–æ–ª—å–∫–æ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π (–∏—Ç–µ—Ä–∞—Ü–∏–π) –Ω–∞–º —Ç—Ä–µ–±—É–µ—Ç—Å—è —Å–¥–µ–ª–∞—Ç—å.

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –∫–æ–¥, –∫–æ—Ç–æ—Ä—ã–π —Ä–∞—Å–ø–µ—á–∞—Ç–∞–µ—Ç 10 —Ä–∞–∑ —Å–ª–æ–≤–æ "–ü—Ä–∏–≤–µ—Ç":

````python
for i in range(10):
    print(i, '–ü—Ä–∏–≤–µ—Ç')
````

–°—Ç—Ä—É–∫—Ç—É—Ä–∞ —Ü–∏–∫–ª–∞ for –≤ Python –≤—ã–≥–ª—è–¥–∏—Ç —Ç–∞–∫:

```
for –Ω–∞–∑–≤–∞–Ω–∏–µ_–ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π_—Ü–∏–∫–ª–∞ in range(–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π):
–±–ª–æ–∫ –∫–æ–¥–∞
```
–î–≤–æ–µ—Ç–æ—á–∏–µ (:) –≤ –∫–æ–Ω—Ü–µ —Å—Ç—Ä–æ–∫–∏ —Å –∏–Ω—Å—Ç—Ä—É–∫—Ü–∏–µ–π for —Å–æ–æ–±—â–∞–µ—Ç –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ç–æ—Ä—É
Python, —á—Ç–æ –¥–∞–ª—å—à–µ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –±–ª–æ–∫ –∫–æ–º–∞–Ω–¥. –í –±–ª–æ–∫ –∫–æ–º–∞–Ω–¥ –≤—Ö–æ–¥—è—Ç –≤—Å–µ —Å—Ç—Ä–æ–∫–∏,
—Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–Ω—ã–µ —Å –æ—Ç—Å—Ç—É–ø–æ–º –æ—Ç —Å—Ç—Ä–æ–∫–∏ —Å –∏–Ω—Å—Ç—Ä—É–∫—Ü–∏–µ–π for, –≤–ø–ª–æ—Ç—å –¥–æ —Å–ª–µ–¥—É—é—â–µ–π —Å—Ç—Ä–æ–∫–∏ –±–µ–∑ –æ—Ç—Å—Ç—É–ø–∞.

–ë–ª–æ–∫ –∫–æ–º–∞–Ω–¥, –∫–æ—Ç–æ—Ä—ã–π –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –≤ —Ü–∏–∫–ª–µ for –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è —Ç–µ–ª–æ–º —Ü–∏–∫–ª–∞.

–í –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö —É—Ä–æ–∫–∞—Ö –º—ã —Å—á–∏—Ç—ã–≤–∞–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ —á–∏—Å–µ–ª, –ø—Ä–∏ –ø–æ–º–æ—â–∏ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –∫–æ–º–∞–Ω–¥ input. –° –ø–æ–º–æ—â—å—é —Ü–∏–∫–ª–∞ for –º–æ–∂–Ω–æ —Å—á–∏—Ç—ã–≤–∞—Ç—å –∏ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å —Å–∫–æ–ª—å–∫–æ —É–≥–æ–¥–Ω–æ —á–∏—Å–µ–ª.

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º —Å–ª–µ–¥—É—é—â–∏–π –ø—Ä–æ–≥—Ä–∞–º–º–Ω—ã–π –∫–æ–¥:

````python

for i in range(5): ]
    num = int(input())
    print('–ö–≤–∞–¥—Ä–∞—Ç –≤–∞—à–µ–≥–æ —á–∏—Å–ª–∞ —Ä–∞–≤–µ–Ω:', num * num)
print('–¶–∏–∫–ª –∑–∞–≤–µ—Ä—à–µ–Ω')
````

–¢–∞–∫–∞—è –ø—Ä–æ–≥—Ä–∞–º–º–∞ —Å—á–∏—Ç—ã–≤–∞–µ—Ç 5 —á–∏—Å–µ–ª –∏ –≤—ã–≤–æ–¥–∏—Ç –Ω–∞ —ç–∫—Ä–∞–Ω –∏—Ö –∫–≤–∞–¥—Ä–∞—Ç—ã –≤–º–µ—Å—Ç–µ —Å –ø–æ—è—Å–Ω—è—é—â–µ–π –Ω–∞–¥–ø–∏—Å—å—é. –ü–æ—Å–∫–æ–ª—å–∫—É –≤—Ç–æ—Ä–∞—è –∏ —Ç—Ä–µ—Ç—å—è —Å—Ç—Ä–æ–∫–∏ –≤—ã–¥–µ–ª–µ–Ω—ã –æ—Ç—Å—Ç—É–ø–æ–º, Python —Å—á–∏—Ç–∞–µ—Ç, —á—Ç–æ —ç—Ç–æ —Ç–µ–ª–æ —Ü–∏–∫–ª–∞, –∫–æ—Ç–æ—Ä–æ–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è 5 —Ä–∞–∑. –ß–µ—Ç–≤–µ—Ä—Ç–∞—è —Å—Ç—Ä–æ–∫–∞ –Ω–µ —Å–æ–¥–µ—Ä–∂–∏—Ç –æ—Ç—Å—Ç—É–ø–∞, –ø–æ—ç—Ç–æ–º—É –Ω–µ —è–≤–ª—è–µ—Ç—Å—è —á–∞—Å—Ç—å—é —Ü–∏–∫–ª–∞ –∏ –±—É–¥–µ—Ç –≤—ã–ø–æ–ª–Ω–µ–Ω–∞ –≤—Å–µ–≥–æ –æ–¥–∏–Ω —Ä–∞–∑, –ø–æ—Å–ª–µ —Ç–æ–≥–æ –∫–∞–∫ —Ü–∏–∫–ª –∑–∞–≤–µ—Ä—à–∏—Ç—Å—è.

–ü—Ä–∏–º–µ—Ä—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Ü–∏–∫–ª–∞ for

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º —Å–ª–µ–¥—É—é—â–∏–π –ø—Ä–æ–≥—Ä–∞–º–º–Ω—ã–π –∫–æ–¥:

````python
print('A')
print('B')
for i in range(5):
    print('C')
    print('D')
print('E')
# A
# B
# C
# C
# C
# C
# C
# D
# D
# D
# D
# D
# E
````

–¢–æ –µ—Å—Ç—å —Å–Ω–∞—á–∞–ª–∞ –ø—Ä–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø–µ—á–∞—Ç–∞–µ—Ç —Å–∏–º–≤–æ–ª—ã –ê –∏ –í, –∑–∞—Ç–µ–º —Å–∏–º–≤–æ–ª—ã C –∏ D –ø—è—Ç—å —Ä–∞–∑, –∞ –∑–∞—Ç–µ–º —Ä–∞—Å–ø–µ—á–∞—Ç–∞–µ—Ç —Å–∏–º–≤–æ–ª –ï –æ–¥–∏–Ω —Ä–∞–∑. –¢–µ–ª–æ —Ü–∏–∫–ª–∞ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –¥–≤—É—Ö —Å—Ç—Ä–æ–∫: —á–µ—Ç–≤–µ—Ä—Ç–æ–π –∏ –ø—è—Ç–æ–π –∏ –∏–º–µ–Ω–Ω–æ –æ–Ω–∏ –±—É–¥—É—Ç –ø–æ–≤—Ç–æ—Ä—è—Ç—å—Å—è.

–í –ø—Ä–æ–≥—Ä–∞–º–º–µ –º–æ–∂–µ—Ç –±—ã—Ç—å —Å–∫–æ–ª—å–∫–æ —É–≥–æ–¥–Ω–æ —Ü–∏–∫–ª–æ–≤. –ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ –º—ã —Ö–æ—Ç–∏–º, —á—Ç–æ–±—ã —Å–Ω–∞—á–∞–ª–∞ 5 —Ä–∞–∑ –±—ã–ª —Ä–∞—Å–ø–µ—á–∞—Ç–∞–Ω —Å–∏–º–≤–æ–ª –°, –∞ –∑–∞—Ç–µ–º 5 —Ä–∞–∑ —Å–∏–º–≤–æ–ª D, –º—ã –º–æ–∂–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å 2 —Ü–∏–∫–ª–∞:

````python
print('A')
print('B')
for i in range(5):
    print('C')
for i in range(5):
    print('D')
print('E')
````

–î–ª—è –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏—è —Ü–∏–∫–ª–∞ –Ω–µ–∫–æ—Ç–æ—Ä–æ–µ –∑–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ —Ä–∞–∑ n –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ü–∏–∫–ª for –≤–º–µ—Å—Ç–µ —Å —Ñ—É–Ω–∫—Ü–∏–µ–π range:



````python
list(range(4))
````

````python
for i in range(4):  # —Ä–∞–≤–Ω–æ—Å–∏–ª—å–Ω–æ –∏–Ω—Å—Ç—Ä—É–∫—Ü–∏–∏ for i in 0, 1, 2, 3:
    # –∑–¥–µ—Å—å –º–æ–∂–Ω–æ –≤—ã–ø–æ–ª–Ω—è—Ç—å —Ü–∏–∫–ª–∏—á–µ—Å–∫–∏–µ –¥–µ–π—Å—Ç–≤–∏—è
    print(i)
    print(i ** 2)
# —Ü–∏–∫–ª –∑–∞–∫–æ–Ω—á–∏–ª—Å—è, –ø–æ—Å–∫–æ–ª—å–∫—É –∑–∞–∫–æ–Ω—á–∏–ª—Å—è –±–ª–æ–∫ —Å –æ—Ç—Å—Ç—É–ø–æ–º
print('–ö–æ–Ω–µ—Ü —Ü–∏–∫–ª–∞')
````

–§—É–Ω–∫—Ü–∏—è range –º–æ–∂–µ—Ç —Ç–∞–∫–∂–µ –ø—Ä–∏–Ω–∏–º–∞—Ç—å –Ω–µ –æ–¥–∏–Ω, –∞ –¥–≤–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞. –í—ã–∑–æ–≤ range(a, b) –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –∏–Ω–¥–µ–∫—Å–Ω–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –±—É–¥–µ—Ç—å –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è –æ—Ç a –¥–æ b - 1, —Ç–æ –µ—Å—Ç—å –ø–µ—Ä–≤—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä —Ñ—É–Ω–∫—Ü–∏–∏ range, –≤—ã–∑—ã–≤–∞–µ–º–æ–π —Å –¥–≤—É–º—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏, –∑–∞–¥–∞–µ—Ç –Ω–∞—á–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–Ω–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π, –∞ –≤—Ç–æ—Ä–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä ‚Äî –ø–µ—Ä–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –∏–Ω–¥–µ–∫—Å–Ω–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –ø—Ä–∏–Ω–∏–º–∞—Ç—å –Ω–µ –±—É–¥–µ—Ç. –ï—Å–ª–∏ –∂–µ a‚â•b, —Ç–æ —Ü–∏–∫–ª –Ω–µ –±—É–¥–µ—Ç –≤—ã–ø–æ–ª–Ω–µ–Ω –Ω–∏ —Ä–∞–∑—É. –ù–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –ø—Ä–æ—Å—É–º–º–∏—Ä–æ–≤–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è —á–∏—Å–µ–ª –æ—Ç 1 –¥–æ n –º–æ–∂–Ω–æ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è —Å–ª–µ–¥—É—é—â–µ–π –ø—Ä–æ–≥—Ä–∞–º–º–æ–π:

````python
summa = 0
n = 5
for i in range(1, n+1):
    summa += i
    print(i, summa)
````

–í —ç—Ç–æ–º –ø—Ä–∏–º–µ—Ä–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è i –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è 1, 2, ..., n, –∏ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π sum –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è –Ω–∞ —É–∫–∞–∑–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.

–ù–∞–∫–æ–Ω–µ—Ü, —á—Ç–æ–±—ã –æ—Ä–≥–∞–Ω–∏–∑–æ–≤–∞—Ç—å —Ü–∏–∫–ª, –≤ –∫–æ—Ç–æ—Ä–æ–º –∏–Ω–¥–µ–∫—Å–Ω–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –±—É–¥–µ—Ç —É–º–µ–Ω—å—à–∞—Ç—å—Å—è, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é range —Å —Ç—Ä–µ–º—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏. –ü–µ—Ä–≤—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä –∑–∞–¥–∞–µ—Ç –Ω–∞—á–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–Ω–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π, –≤—Ç–æ—Ä–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä ‚Äî –∑–Ω–∞—á–µ–Ω–∏–µ, –¥–æ –∫–æ—Ç–æ—Ä–æ–≥–æ –±—É–¥–µ—Ç –∏–∑–º–µ–Ω—è—Ç—å—Å—è –∏–Ω–¥–µ–∫—Å–Ω–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è (–Ω–µ –≤–∫–ª—é—á–∞—è –µ–≥–æ!), –∞ —Ç—Ä–µ—Ç–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä ‚Äî –≤–µ–ª–∏—á–∏–Ω—É –∏–∑–º–µ–Ω–µ–Ω–∏—è –∏–Ω–¥–µ–∫—Å–Ω–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π. –ù–∞–ø—Ä–∏–º–µ—Ä, —Å–¥–µ–ª–∞—Ç—å —Ü–∏–∫–ª –ø–æ –≤—Å–µ–º –Ω–µ—á–µ—Ç–Ω—ã–º —á–∏—Å–ª–∞–º –æ—Ç 1 –¥–æ 99 –º–æ–∂–Ω–æ –ø—Ä–∏ –ø–æ–º–æ—â–∏ —Ñ—É–Ω–∫—Ü–∏–∏ range(1, 100, 2), –∞ —Å–¥–µ–ª–∞—Ç—å —Ü–∏–∫–ª –ø–æ –≤—Å–µ–º —á–∏—Å–ª–∞–º –æ—Ç 100 –¥–æ 1 –º–æ–∂–Ω–æ –ø—Ä–∏ –ø–æ–º–æ—â–∏ range(100, 0, -1).

–ë–æ–ª–µ–µ —Ñ–æ—Ä–º–∞–ª—å–Ω–æ, —Ü–∏–∫–ª ```for i in range(a, b, d) –ø—Ä–∏ d > 0``` –∑–∞–¥–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è –∏–Ω–¥–µ–∫—Å–Ω–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ```i = a, i = a + d, i = a + 2 * d``` –∏ —Ç–∞–∫ –¥–ª—è –≤—Å–µ—Ö –∑–Ω–∞—á–µ–Ω–∏–π, –¥–ª—è –∫–æ—Ç–æ—Ä—ã—Ö ```i < b```. –ï—Å–ª–∏ –∂–µ ```d < 0```, —Ç–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è —Ü–∏–∫–ª–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è ```i > b```.

````python
print(list(range(5)))
print(list(range(1,10)))
print(list(range(1, 15, 4)))
````

–í–æ–∑—å–º–∏ —ç–ª–µ–º–µ–Ω—Ç—ã —Å i-–≥–æ (–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ) –ø–æ j-—ã–π (–Ω–µ–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ) —Å —à–∞–≥–æ–º k

````python
a = [1, 2, 3, 3.1415]
a.append(4) #–µ—Å–ª–∏ —Ö–æ—Ç–∏–º –¥–æ–±–∞–≤–∏—Ç—å —Ç–æ–ª—å–∫–æ 1 —ç–ª–µ–º–µ–Ω—Ç
print(a)
a.extend([1, 2, 3]) # –µ—Å–ª–∏ —Ö–æ—Ç–∏–º –¥–æ–±–∞–≤–∏—Ç—å —Å—Ä–∞–∑—É –Ω–µ—Å–∫–æ–ª—å–∫–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤
print(a)
print(a + [1, 2, 3]) # –∞ –µ—â–µ –º–æ–∂–Ω–æ –≤–æ—Ç —Ç–∞–∫
````

–ê –µ—â–µ –º–æ–∂–Ω–æ –ø—Ä–æ–¥—É–±–ª–∏—Ä–æ–≤–∞—Ç—å —Å –ø–æ–º–æ—â—å—é —É–º–Ω–æ–∂–µ–Ω–∏—è!

````python
a * 3
````

````python
a.insert(1, -15)
print(a)
````

–ß—Ç–æ –µ—â–µ —É–º–µ–µ–º –¥–µ–ª–∞—Ç—å —Å–æ —Å–ø–∏—Å–∫–∞–º–∏?

* len(l) - –≤—ã–≤–µ—Å—Ç–∏ —Ä–∞–∑–º–µ—Ä —Å–ø–∏—Å–∫–∞ (—Å–∫–æ–ª—å–∫–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–æ–¥–µ—Ä–∂–∏—Ç)

* count(elem) - –ø–æ—Å—á–∏—Ç–∞–π, —Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ —ç–ª–µ–º–µ–Ω—Ç –≤—Ö–æ–¥–∏—Ç –≤ —Å–ø–∏—Å–æ–∫

* index(elem) - –≤—ã–≤–µ–¥–∏ –∏–Ω–¥–µ–∫—Å –ø–µ—Ä–≤–æ–≥–æ –≤—Ö–æ–∂–¥–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞ (–µ—Å–ª–∏ —ç–ª–µ–º–µ–Ω—Ç–∞ –Ω–µ—Ç, —Ç–æ –≤—ã–≤–µ–¥–µ—Ç –æ—à–∏–±–∫—É)

* reverse() - –ø–µ—Ä–µ–≤–µ—Ä–Ω–∏ —Å–ø–∏—Å–æ–∫

* sort(reverse=False) - –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–π —Å–ø–∏—Å–æ–∫ (–ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é, –µ—Å–ª–∏ –ø–æ—Å—Ç–∞–≤–∏—Ç—å reverse=True, —Ç–æ –±—É–¥–µ—Ç –ø–æ —É–±—ã–≤–∞–Ω–∏—é). –ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ –µ—Å—Ç—å —Ñ—É–Ω–∫—Ü–∏—è sorted(), –∫–æ—Ç–æ—Ä–∞—è –¥–µ–ª–∞–µ—Ç —Ç–æ –∂–µ —Å–∞–º–æ–µ, –Ω–æ —Ç–æ–ª—å–∫–æ –Ω–µ –º–µ–Ω—è–µ—Ç —Å–ø–∏—Å–æ–∫

* max/min(a) - –Ω–∞–π—Ç–∏ –º–∞–∫—Å–∏–º—É–º/–º–∏–Ω–∏–º—É–º –≤ —Å–ø–∏—Å–∫–µ

````python
a = [1, 2, 3, 3.1415, 1, 2, 3]
print(len(a))
print(a.count(2))
a.reverse()
print(a)
a.sort()
print(sorted(a, reverse=True), a)
````

–ú–µ—Ç–æ–¥–æ–≤ –∏ —Ñ—É–Ω–∫—Ü–∏–π —É string –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –º–Ω–æ–≥–æ (–º–æ–∂–Ω–æ –≤—Å–µ –Ω–∞–π—Ç–∏ [–≤–æ—Ç —Ç—É—Ç](https://docs.python.org/3/library/stdtypes.html#string-methods)), –ø–µ—Ä–µ—á–∏—Å–ª–∏–º —Ç–æ–ª—å–∫–æ —Å–∞–º—ã–µ –≤–∞–∂–Ω—ã–µ –∏–∑ –Ω–∏—Ö:

* len(s) - –≤—ã–≤–µ–¥–∏ –¥–ª–∏–Ω—É —Å—Ç—Ä–æ–∫–∏

* s.isdigit() - –ø—Ä–æ–≤–µ—Ä–∫–∞, —è–≤–ª—è—é—Ç—Å—è –ª–∏ –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã –≤ —Å—Ç—Ä–æ–∫–µ —á–∏—Å–ª–∞–º–∏ (–º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –∫–æ–Ω–≤–µ—Ä—Ç–∞—Ü–∏–∏ —Å—Ç—Ä–æ–∫–∏ –≤ —á–∏—Å–ª–æ)

* s.islower(), s.isupper() - –ø—Ä–æ–≤–µ—Ä–∫–∞, —è–≤–ª—è—é—Ç—Å—è –ª–∏ –≤—Å–µ –±—É–∫–≤—ã –∑–∞–≥–ª–∞–≤–Ω—ã–º–∏ –∏–ª–∏ —Å—Ç—Ä–æ—á–Ω—ã–º–∏

* join() - —Å–æ–µ–¥–∏–Ω–∏—Ç—å —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫ –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É
