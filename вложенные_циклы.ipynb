```python
if a[0]>a[1]:
  a[0],a[1] = a[1],a[0]

for i in range(0, len(a)-1):
  print(f'Индекс i = {i}, элементы: {a[i]}, {a[i+1]}')
  if a[i]>a[i+1]:
    a[i],a[i+1] = a[i+1],a[i]
  print(f'Индекс i = {i}, элементы: {a[i]}, {a[i+1]}')
  print(a)
```

```python
b = [21, 20, 19, 18, 17, 16]
for i in range(len(b)):
  print('i =', i)
  for j in range(0, len(b)-1):
    print(f'Индекс j = {j}, элементы: {b[j]}, {b[j+1]}')
    if b[j]>b[j+1]:
      b[j],b[j+1] = b[j+1],b[j]
    print(f'Индекс j = {j}, элементы: {b[j]}, {b[j+1]}')
    print(b)
```
